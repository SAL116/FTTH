```python
import pandas as pd
data = pd.read_csv('final.csv')
```

```python
import numpy as np
```

```python
temp = pd.read_csv('all_states3.csv')
temp.head()
```

```python
data=data.fillna(0)
data.head()
```

```python
df = data[data.down14_fiber==0]
df.head()
```

```python
df['change']=df.down15_fiber-df.down14_fiber
```

```python
df.head()
```

```python
def f(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
df.change=df.change.apply(f)
```

```python
df.head()
```

```python
x = df.down14_cable
y = np.floor(np.log(df.POP10.values))
df['providers14']=x
df['log_population14']=y
```

```python
import matplotlib 
import matplotlib.pyplot as plt
import math
import numpy as np
%matplotlib inline
x = np.ceil(df.down14_total)
y = np.floor(np.log(df.POP10.values))
df['providers14']=x
df['log_population14']=y

colors = np.where(df["change"]==1,'r','blue')
df.plot.scatter(x='providers14',y='log_population14',c=colors,alpha=1,figsize=(8,8))

plt.xlabel('Number of Providers')
plt.ylabel('Log number of Population')
plt.title('Fiber change 2014-2015')
red = matplotlib.lines.Line2D(range(1), range(1), color="white", marker='o', markerfacecolor="red")
blue = matplotlib.lines.Line2D(range(1), range(1), color="white", marker='o', markerfacecolor="blue")
handles = [red,blue]
labels = ['Using Fiber','No Fiber']
plt.legend(handles=handles, labels=labels,loc='lower right',bbox_to_anchor=(1.2, 0.5),bbox_transform=plt.gcf().transFigure)
plt.savefig("1.png", bbox_inches="tight")
plt.show()
```
