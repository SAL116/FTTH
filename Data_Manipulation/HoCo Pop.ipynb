```python
def diff(df,y1,y2):
    df3 = df[df.year == y1].merge(df[df.year == y2],how = 'outer',on = ['HoldingCompanyName','TechType'])
    df3['diff'] = df3.population_y - df3.population_x
    df3 = df3[df3.TechType=='fiber'].merge(df3[df3.TechType=='cable'],on = 'HoldingCompanyName',how='left').merge(df3[df3.TechType=='dsl'],on = 'HoldingCompanyName',how='left')
    df3.rename(columns = {'diff_x':'fiber','diff_y':'cable','diff':'dsl'},inplace=True)
    return df3[['HoldingCompanyName','fiber','cable','dsl']].sort_values('fiber',ascending=False)
```
